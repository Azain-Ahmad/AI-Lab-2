```python
# Simple Hill Climbing Algorithm

def hill_climbing(arr):

    best = arr[0]

    for value in arr:

        # Move to better value
        if value > best:
            best = value

    return best


values = [3, 7, 12, 5, 20, 15]

result = hill_climbing(values)

print("Best Value:", result)
```

1. What is Hill Climbing algorithm?
Hill Climbing is a heuristic search algorithm used in artificial intelligence to find the best solution by continuously moving toward better states.

2. Is Hill Climbing a heuristic search or blind search method?
It is a heuristic search method because it uses an evaluation function to guide the search.

3. What is the main idea behind Hill Climbing?
The main idea is:
 Start with an initial solution and repeatedly improve it by moving to a better neighboring solution until no improvement is possible.

4. Why is it called a “greedy” algorithm?
Because it always chooses the best immediate (local) option without considering the future consequences.

5. What type of problems is Hill Climbing used for?
It is used for optimization problems, such as:

Route finding
Scheduling
Puzzle solving
Machine learning optimization
 Working of Algorithm

6. How does Hill Climbing choose the next state?
It selects the neighbor state that gives a better value according to the evaluation function.

7. What is a “current state” in Hill Climbing?
It is the present solution or position the algorithm is currently at.

8. What is a “neighbor state”?
A neighbor state is a solution that is slightly different from the current state.

9. What is the role of the evaluation function?
It measures how good a state is and helps decide which state is better.

10. What happens if no better neighbor is found?
The algorithm stops because it assumes a local optimum has been reached.

 Types and Variations

11. What are the types of Hill Climbing algorithms?

Simple Hill Climbing
Steepest Ascent Hill Climbing
Stochastic Hill Climbing

12. What is Simple Hill Climbing?
It checks neighbors one by one and moves to the first better state found.

13. What is Steepest Ascent Hill Climbing?
It checks all neighbors and selects the best possible improvement.

14. What is Stochastic Hill Climbing?
It randomly selects a better neighbor instead of checking all options.

15. How does Steepest Ascent differ from Simple Hill Climbing?

Simple: chooses first better neighbor
Steepest: chooses the best among all neighbors
 Advantages and Disadvantages

16. What are the advantages of Hill Climbing?

Simple to implement
Requires less memory
Works well for large search spaces
Fast in many cases

17. What are the disadvantages of Hill Climbing?

Can get stuck in local maxima
No backtracking
May miss global optimum
Sensitive to starting point

18. What is the problem of local maxima?
It is a state where a solution is better than all its neighbors but not the best overall solution.

19. What is a plateau in Hill Climbing?
A flat area where all neighboring states have the same value, so the algorithm cannot decide where to move.

20. What is a ridge problem in Hill Climbing?
A ridge is a narrow path of improvement that is difficult for the algorithm to follow because it cannot move directly upward.
